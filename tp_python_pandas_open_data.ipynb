## Step 1:
Download the dataset to a local .csv file using python library request


```python
req = requests.get('https://parisdata.opendatasoft.com/explore/dataset/coronavirus-commercants-parisiens-livraison-a-domicile/download/?format=csv&timezone=Europe/Berlin&lang=fr&use_labels_for_header=true&csv_separator=%3B', stream=True)
with open("https://parisdata.opendatasoft.com/explore/dataset/coronavirus-commercants-parisiens-livraison-a-domicile/download/?format=csv&timezone=Europe/Berlin&lang=fr&use_labels_for_header=true&csv_separator=%3B", 'r') as fp:
  for partie in req.iter_content:
        fp.write(partie)
```
